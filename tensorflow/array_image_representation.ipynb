# Array image representation

As the machine learning processes data as multi-dimentional array, all the data(training, test, etc) must be converted into multi-dimentional array. The code shows how to process image data into array by using PIL(Pillo) the most famous and popular library for image processing. 

As RGB(Red, Green, Blue) model is the most popular representation of the image in a digital way, image is converted into 3 dimentional array that represents each color. 

[RGB Color Model](https://en.wikipedia.org/wiki/RGB_color_model)

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load .jpeg image. 
im = Image.open("../images/back_to_the_future.jpeg")

# You can display the image with matplotlib.imshow()
# In this phase, the image data is still in raw data. 
plt.imshow(im)
```

```python
# Convert into ndarray with numpy. 
a = np.asarray(im)

# The image data is now in 3 dimentional array. 
print(a.shape)
```

```python
# You also can take array data as an argument of matplolib.imshow().
plt.imshow(a)
plt.show()
```
